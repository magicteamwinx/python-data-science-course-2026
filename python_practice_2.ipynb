Список (list) — це динамічний контейнер, який є змінним (mutable), що дозволяє вільно редагувати, додавати або видаляти елементи під час роботи програми. Він ідеально підходить для зберігання колекцій, склад яких заздалегідь невідомий, наприклад, списку користувачів у чаті.

Кортеж (tuple) — це статична структура, яка є незмінною (immutable); після створення її вміст зафіксовано, що гарантує цілісність даних і захищає від випадкових помилок. Завдяки цій незмінності кортежі працюють швидше, займають менше місця в пам'яті та можуть використовуватися як ключі у словниках, що робить їх кращим вибором для констант або чітких структур, як-от координати [x,y].

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
unique_items = set()

unique_items.add("Apple")
unique_items.add(42)
unique_items.add(3.14)

print(f"Вміст множини: {unique_items}")
```

Так, множина (set) є змінним (mutable) типом даних. Ви можете додавати нові елементи (.add()), видаляти існуючі (.remove(), .discard()) або об'єднувати її з іншими множинами. Однак є важливе обмеження: самі елементи всередині множини повинні бути незмінними (immutable), наприклад, числа, рядки або кортежі. Ви не зможете покласти список всередину множини.

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
mixed_dict = {}

mixed_dict[100] = "Integer key"          # Ключ - int
mixed_dict["name"] = "Anastasiia"        # Ключ - str
mixed_dict[3.14] = "Float key"           # Ключ - float
mixed_dict[(1, 2)] = "Tuple key"         # Ключ - tuple

for key in mixed_dict.keys():
    print(f"Key: {key} | Type: {type(key)}")
```

Ні, список (list) не може бути ключем словника. Це пояснюється тим, що список — змінний (mutable) тип даних. Для роботи словника ключі повинні бути хешованими, тобто їхнє значення та унікальний "відбиток" (hash) не повинні змінюватися протягом усього часу існування. Оскільки вміст списку можна редагувати, він не має стабільного хешу.
